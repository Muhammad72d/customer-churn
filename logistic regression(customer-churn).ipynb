```python
#بسم الله 
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
from joblib import dump
```

```python
data=pd.read_excel('Telco_customer_churn.xlsx')
```

```python
le=LabelEncoder()
```

```python
print(data.columns)
```

```python
for i in data.columns:
    if not data[i].dtypes =='int' and not data[i].dtypes =='float' and not data[i].dtypes =='string' :
        data[i]='not found'  
    data=data[data!='not found'] 
    if not data[i].dtypes =='int' and not data[i].dtypes =='float':
       data[i]=le.fit_transform(data[i])
```

```python
print(data.dtypes)
```

```python
print(data['Total Charges'])
```

```python
y=data['Churn Value']
data=data.drop(columns='Churn Value')
```

```python
xt,xs,yt,ys=train_test_split(data,y,test_size=0.2,random_state=41)
```

```python
model=make_pipeline(StandardScaler(),LogisticRegression())
#pipline(steps=[('standardscaler',SrandardScaler()),('logisticregression',LogisticRegression())])
```

```python
model.fit(xt,yt)
```

pred=model.predict(xt)

```python
pred=model.predict(xs)
```

```python
print(yt[:10])
print(pred[:10])
```

```python
error=mean_squared_error(ys,pred)
```

```python
print(error)
```

```python
dump(model,'trained logistic regression model')
```
